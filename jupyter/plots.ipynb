```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
ml_df = pd.read_csv("results/ML.csv")
ml_pso_df = pd.read_csv("results/ML_PSO.csv")
ml_gow_df = pd.read_csv("results/ML_GOW.csv")
ml_woa_df = pd.read_csv("results/ML_WOA.csv")
ml_pca_df = pd.read_csv("results/ML_PCA.csv")
ml_pca_pso_df = pd.read_csv("results/ML_PCA_PSO.csv")
ml_pca_gow_df = pd.read_csv("results/ML_PCA_GOW.csv")
ml_pca_woa_df = pd.read_csv("results/ML_PCA_WOA.csv")

# Combine the dataframes into a single dataframe for easier plotting
ml_df["Method"] = "ML"
ml_pso_df["Method"] = "ML_PSO"
ml_gow_df["Method"] = "ML_GOW"
ml_woa_df["Method"] = "ML_WOA"
ml_pca_df["Method"] = "ML_PCA"
ml_pca_pso_df["Method"] = "ML_PCA_PSO"
ml_pca_gow_df["Method"] = "ML_PCA_GOW"
ml_pca_woa_df["Method"] = "ML_PCA_WOA"

combined_df = pd.concat([ml_df, ml_pso_df, ml_gow_df, ml_woa_df,ml_pca_df, ml_pca_pso_df, ml_pca_gow_df, ml_pca_woa_df])

# Plotting the bar plot comparing models
plt.figure(figsize=(14, 8))
model_comparison = combined_df.pivot(
    index="Classifier", columns="Method", values="Accuracy"
)
model_comparison.plot(kind="bar", figsize=(14, 8))
plt.title("Model Comparison Across Different Methods")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.legend(title="Method")
plt.show()

# Plotting the bar plot comparing methods
plt.figure(figsize=(14, 8))
method_comparison = combined_df.pivot(
    index="Method", columns="Classifier", values="Accuracy"
)
method_comparison.plot(kind="bar", figsize=(14, 8))
plt.title("Method Comparison Across Different Models")
plt.ylabel("Accuracy")
plt.xticks(rotation=45)
plt.legend(title="Classifier")
plt.show()
```
